```python
import random, time

ROWS = int(input("Hang: "))
COLS = int(input("Cot : "))

room = []
print(f"Nhap tung hang ({COLS} so: 0=sach | 1,2,3=ban | -1=vat can)")
for r in range(ROWS):
    while True:
        try:
            row = list(map(int, input(f"  Hang {r+1}: ").split()))
        except:
            print("  Chi nhap so. Nhap lai."); continue
        if len(row) != COLS:
            print(f"  Can dung {COLS} so. Nhap lai."); continue
        if any(v < -1 for v in row):
            print("  Gia tri nho nhat la -1. Nhap lai."); continue
        room.append(row)
        break
print("Room da nhap:"); [print(f"  {r}") for r in room]
```

```python
MOVE = {"UP":(-1,0),"DOWN":(1,0),"LEFT":(0,-1),"RIGHT":(0,1)}

def build_rules(grid, rows, cols):
    rules = {}
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == -1: continue
            moves = []
            for action, dr, dc in [("UP",-1,0),("DOWN",1,0),("LEFT",0,-1),("RIGHT",0,1)]:
                nr, nc = r+dr, c+dc
                if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] != -1:
                    moves.append(action)
            rules[(r,c)] = moves
    return rules

def show(grid, pos):
    print("  +" + "---+"*len(grid[0]))
    for r in range(len(grid)):
        row_str = "  |"
        for c in range(len(grid[0])):
            if (r,c)==pos:        row_str += " R |"
            elif grid[r][c]==-1:  row_str += " # |"
            elif grid[r][c]==0:   row_str += " . |"
            else:                 row_str += f" {grid[r][c]} |"
        print(row_str)
        print("  +" + "---+"*len(grid[0]))

def clean(grid, r, c):
    if grid[r][c] > 0: grid[r][c] -= 1

def move(pos, action, grid):
    dr, dc = MOVE[action]
    nr, nc = pos[0]+dr, pos[1]+dc
    if 0 <= nr < len(grid) and 0 <= nc < len(grid[0]) and grid[nr][nc] != -1:
        return (nr, nc)
    return pos

def all_clean(grid):
    return all(grid[r][c] <= 0
               for r in range(len(grid))
               for c in range(len(grid[0]))
               if grid[r][c] != -1)

rules = build_rules(room, ROWS, COLS)
free  = [(r,c) for r in range(ROWS) for c in range(COLS) if room[r][c] != -1]
print(f"O hop le: {free}")

while True:
    try:
        r0, c0 = map(int, input("Vi tri bat dau (r c): ").split())
        if (r0,c0) in free: pos = (r0,c0); break
        print("  Khong hop le. Nhap lai.")
    except:
        print("  Nhap dang: r c")

import copy
room_backup = copy.deepcopy(room)
print(f"\nStart | Pos {pos}"); show(room, pos)
```

```python
room = copy.deepcopy(room_backup)
MAX  = int(input("So buoc toi da: "))

for step in range(1, MAX+1):
    r, c    = pos
    percept = room[r][c]

    if percept > 0:
        action = "CLEAN"
        clean(room, r, c)
    else:
        valid  = rules.get((r,c), [])
        action = random.choice(valid) if valid else "STAY"
        if action != "STAY":
            pos = move(pos, action, room)

    print(f"\nStep {step:02d} | Pos ({r},{c}) | Percept: {percept} | Action: {action}")
    show(room, pos)
    time.sleep(0.4)

    if all_clean(room):
        print(f"\nDon sach sau {step} buoc!"); break
else:
    print(f"\nHet {MAX} buoc.")
```
